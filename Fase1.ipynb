```python
def plota_grafico_personalizado(dados, meses_selecionados, figsize=(20, 6)):

    fig, ax = plt.subplots(figsize=figsize)
    dados.T.plot(ax=ax)

    ax.set_xticks(range(len(meses_selecionados)))
    ax.set_xticklabels(meses_selecionados, rotation=45)

    ax.set_ylabel("Gasto aprovado em milhões de reais")
    ax.set_xlabel("Anos")

    plt.show()
```

```python
def plota_gastos_por_mes(dados, figsize=(10,6)): #cria uma função 
    axis = dados.T.plot(figsize=figsize)
    axis.set_ylabel("Gasto aprovado em milhões de reais")
    axis.set_xlabel("mês")
```
